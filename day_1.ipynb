```python
import pandas as pd
df = pd.read_csv('./student_placement_dataset.csv') #loading the dataset

X = df.iloc[:, 0:2] #selecting the first two columns as features
y = df.iloc[:, -1] #selecting the last column as target variable
```

```python
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split #importing train test split for splitting the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler #importing standard scaler for feature scaling
scaler = StandardScaler() #creating an object of standard scaler
X_train = scaler.fit_transform(X_train) #transforming the training data to an array
X_train
```

```python
X_test=scaler.transform(X_test) #transforming the test data to an array 
X_test
```

```python
from sklearn.linear_model import LogisticRegression #importing logistic regression model
clr=LogisticRegression() #creating an object of logistic regression model
clr.fit(X_train,y_train)#training the model
```

```python
y_pred=clr.predict(X_test)#model prediction
```

```python
y_test#actual values
```

```python
from sklearn.metrics import accuracy_score #accuracy of the model
accuracy_score(y_test,y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions #visualization of decision boundary
plot_decision_regions(X_train, y_train.values, clf=clr,legend=2) 
```
